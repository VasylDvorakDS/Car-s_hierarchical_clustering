```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model=AgglomerativeClustering(n_clusters=3)
scaled_df
```

```python
cluster_labels=model.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
plt.figure(figsize=(12, 4), dpi=200)
sns.scatterplot(data=df, x='mpg', y='weight', hue=cluster_labels);
```

```python
plt.figure(figsize=(12, 4), dpi=200)
sns.scatterplot(data=df, x='mpg', y='horsepower', hue=cluster_labels, palette='viridis');
```

```python
model= AgglomerativeClustering(n_clusters=None, distance_threshold=2)
```

```python
cluster_labels=model.fit_predict(scaled_df)
cluster_labels
```

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
```

```python
linkege_matrix=hierarchy.linkage(model.children_)
linkege_matrix
```

```python
plt.figure(figsize=(20, 10), dpi=200)
dendro=dendrogram(linkege_matrix, truncate_mode='lastp', p=11)
```

```python
plt.figure(figsize=(20, 10), dpi=200)
dendro=dendrogram(linkege_matrix, truncate_mode='level', p=4)
```
